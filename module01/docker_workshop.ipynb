We need  both <b>python</b> and <b>docker</b> installed. However, both of these are included in codespaces, so no need to install locally. We will simply use Github Codespaces!

# 1. What is Docker?

- Docker is a way to separate the things on your host machine from the things on the applications that you create.
- docker creates this isolated container that specifies the host you want to use(i.e. Linux Ubuntu 2.4)

## 2. What is a data pipeline?

- A data pipeline is a system that takes data as input and produces more data as output
- Ex. a script that takes data from a CSV file and outputs it to a Postgres database

## 3. Our first data pipeline: a python script

- in our first data pipeline we will be writing a python script
- but how do you allow your script to take in  parameters:
    - we import the **sys** module, which will give us access to the arguments passed in when the script is ran
    - the arguments will be made availabe in the sys.argv attribute which holds the arguments as a list
